### Importing Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

### Reading the data

```python
train_df= pd.read_csv('./input/leaf-classification/train.csv.zip')
test_df=pd.read_csv('./input/leaf-classification/test.csv.zip')
```

### Machine learning

```python
X = train_df.drop(['id','species'],axis=1)
Y = train_df['species']
```

```python
encoder = LabelEncoder()
y_fit = encoder.fit(train_df['species'])
y_label = y_fit.transform(train_df['species'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y_label, test_size = 0.2, random_state =1)
```

### Neural Networks

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.fit_transform(x_test)

mlp = MLPClassifier(max_iter=1000, alpha=1, random_state=0)
mlp.fit(X_train_scaled, y_train)

print("Accuracy on training set: {:.3f}".format(
    mlp.score(X_train_scaled, y_train)))
print("Accuracy on test set: {:.3f}".format(mlp.score(X_test_scaled, y_test)))
```
